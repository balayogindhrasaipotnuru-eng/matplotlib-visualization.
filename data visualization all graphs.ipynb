```python
import  matplotlib.pyplot as plt


cars =['BMW','MERCEDES','AUDI','TOYOTA','HONDA',]
speed = [200,240,150,225,220]

plt.pie (speed, labels=cars, autopct='%1.1f%%')
plt.title('pie chart')
plt.show()

```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [2,3,5,7,11]

plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('line plot by balayogi')
plt.show()
```

```python

import matplotlib.pyplot as plt 
name = ["A","B","C","D","E"]
value =[10,20,30,40,50]
plt.bar(name,value)
plt.xlabel("name ")
plt.ylabel("value")
plt.title("Bar graph by balayogi")
plt.show()
```

```python
import matplotlib.pyplot as plt
v1 = [1,2,3,4,5]
v2 = [10,20,30,40,50]
plt.scatter(v1,v2)
plt.xlabel("v1")
plt.ylabel("v2")
plt.title("Scatter plot by balayogi")
plt.show()
```

```python
import matplotlib.pyplot as plt 
something = [10,20,30,40,50]

plt.hist(something, bins=5)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram by balayogi")
plt.show()
```

```python
import matplotlib.pyplot as plt 
something = [10,20,30,40,50]
plt.boxplot(something)
plt.title ("box plot by balayogi")
plt.show()
```
